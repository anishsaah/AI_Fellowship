# Label based selection

```
# data frame is created to load csv on the next cell.

import pandas as pd
data = {'Company': ['Ford', 'Ferrari', 'Lamborghini', 'Toyota', 'Ford', 'Toyota'],
        'Color': ['red', 'blue', 'white', 'white', 'blue', 'red'],
        'Doors': [4, 2, 2, 4, 4, 2]

        }
pd.DataFrame(data).to_csv("car_file.csv", index=False)
```

```
# To under a label based selection in dataframe, assume, we have a csv file containing three columns and six rows as shown here. 
# We read that csv file with index column as company. Here, in read csv function, parameter index column allow you to set which columns to be used as the index of the dataframe. 

df = pd.read_csv('/content/car_file.csv', index_col='Company')
df
```

```
# Now, we use loc method with index operator with name of the label inside it, in this case label is ford. So, by running this line of code returns all the rows with label ford as shown here.
df.loc]"Ford"]
```

# Index based selection

```
# To explain the index based selection in dataframe, assume, we have a dataframe named df containing three columns and six rows as shown here. 

import pandas as pd
data = {'Company': ['Ford', 'Ferrari', 'Lamborghini', 'Toyota', 'Ford', 'Toyota'],
        'Color': ['red', 'blue', 'white', 'white', 'blue', 'red'],
        'Doors': [4, 2, 2, 4, 4, 2]

        }
df = pd.DataFrame(data)
```

```
# Now, we use iloc method with index operator with a particular positions of the index inside it. 
# In this case integer two is placed. So, by running this line of code retrieves row from this dataframe as you can see row of index label two has been returned.

df.iloc[2]
```

```
# Similarly, we can also use iloc method to return a dataframe with multiple rows. By running this line of code we can see, rows with index label 2 and 3 have been returned. 

df.iloc[2:4]
```
